```python
import pandas as pd
import json
```

```python
SDA1 = json.load(open("Senior_Data_Analyst_Resumes.json"))
SDA1 = pd.DataFrame.from_dict(SDA1["position"])
SDA1.columns = ["company", "description", "location", "rating", "rating_url", "position"]
SDA2 = json.load(open("Senior_Data_Analyst_Resumes2.json"))
SDA2 = pd.DataFrame.from_dict(SDA2["position"])
SDA2.columns = ["position", "company", "description", "location", "rating", "rating_url"]
DS = json.load(open("Data_Scientist.json"))
DS = pd.DataFrame.from_dict(DS["position"])
DS.columns = ["position", "company", "description", "location", "rating", "rating_url"]
DS2 = json.load(open("Data_Scientist2.json"))
DS2 = pd.DataFrame.from_dict(DS2["position"])
DS2.columns = ["position", "company", "description", "location", "rating", "rating_url"]
```

```python
all_data = pd.concat([SDA1, SDA2, DS, DS2], sort = True).drop_duplicates(subset = ["position", "company", "location", "rating"])
all_data = all_data.drop(columns = ["rating_url"])
all_data = all_data.dropna()
```

```python
all_data["description"] = all_data["description"].str.replace("\\n", " ")
all_data["description"] = all_data.apply(lambda row : row["description"].replace(str(row["location"]), ''), axis=1)
all_data
```

```python
all_data.to_csv('resume_list.csv')
```
